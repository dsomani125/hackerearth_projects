```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
```

```python
test_set.head()
```

```python
test_copy = test_set.copy()
```

```python
train_set.drop('Employee_ID', inplace = True , axis = 1)
test_set.drop('Employee_ID', inplace = True , axis = 1)
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
x = train_set.iloc[:,:-1].values
y = train_set.iloc[:,-1].values
x_test =  test_set.iloc[:,:].values
```

```python
print(x_test[0])
```

```python
print(x[0])
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan , strategy = 'mean')
imputer.fit(x[:,[1,2,7,8,9,10,11,12,14,15,16,17,18,19,20,21]])
x[:,[1,2,7,8,9,10,11,12,14,15,16,17,18,19,20,21]] = imputer.transform(x[:,[1,2,7,8,9,10,11,12,14,15,16,17,18,19,20,21]])
x_test[:,[1,2,7,8,9,10,11,12,14,15,16,17,18,19,20,21]] = imputer.transform(x_test[:,[1,2,7,8,9,10,11,12,14,15,16,17,18,19,20,21]])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
le4 = LabelEncoder()
le5 = LabelEncoder()
x[:,0] = le.fit_transform(x[:,0])
x_test[:,0] = le.transform(x_test[:,0])
x[:,3] = le1.fit_transform(x[:,3])
x_test[:,3] = le1.transform(x_test[:,3])
x[:,4] = le2.fit_transform(x[:,4])
x_test[:,4] = le2.transform(x_test[:,4])
x[:,5] = le3.fit_transform(x[:,5])
x_test[:,5] = le3.transform(x_test[:,5])
x[:,6] = le4.fit_transform(x[:,6])
x_test[:,6] = le4.transform(x_test[:,6])
x[:,13] = le5.fit_transform(x[:,13])
x_test[:,13] = le5.transform(x_test[:,13])
```

```python
print(x[0])
```

```python
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x_test = sc.transform(x_test)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test1 , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(x, y)
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x,y)
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x, y)
y_pred = regressor.predict(x_test)
print(y_pred)
```

```python
final_answer = pd.DataFrame({
    'Employee_ID' : test_copy['Employee_ID'],
    'Attrition_rate' : y_pred
})
```
